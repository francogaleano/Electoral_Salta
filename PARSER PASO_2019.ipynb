```python
#IMPORTAMOS LOS PAQUETES
import pandas as pd
import glob
```

```python
#CARGAMOS TODAS LAS BASES DE DATOS
path = r'ciudades' 
all_files = glob.glob(path + "/*.csv")

lista_archivos = []

for nombre_archivo in all_files:
    df = pd.read_csv(nombre_archivo, index_col=None, header=0)
    lista_archivos.append(df)

frame = pd.concat(lista_archivos, axis=0, ignore_index=True)
```

### Como se puede observar tenemos bases al estilo "LONG", para poder procesar los datos vamos a pasar las bases al estilo "wide". De modo que cada columna sea los votos de una fuerza pol√≠tica y la celda sea la cantidad de votos que obtuvo

```python
#TENEMOS BASES AL ESTILO LONG
frame.head(15)
```

### PASAMOS LOS DATOS A FORMATO "WIDE"
